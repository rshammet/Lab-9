Instead of discussing the information about its phylogenetic relatives, the evolutionary relationships around bacteria will be discussed. As shown throughout this lab, the basis of these evolutionary relationships is the 16 S rRNA sequences. From this, 11 main groups (or divisions) among the Bacteria were proposed (1). These groups were created on the basis of the phylogenetic trees, they were all separated by long naked branches. As our information on Bacteria has expanded so has the amount of groups/divisions found. However, it is still unclear whether the new groups found are separate from the original 11 groups or if they are subcategories from within. Due to this, new ways of how to find shared ancestors and find the links to the 16 S rRNA sequences are constantly being explored. One way is finding use of the conserved inserts and deletions found in various proteins (1). This approach provides an accurate way to place different bacterial genomes into their respected groups along the phylogenetic placement. 

The relatives of the bacterium are from the Pseudomonas genome, as stated above. They live in aquatic environments and mostly tropical/temperate climate conditions (2). It can also be consumed by a host so that is another environment that should be kept in mind. Due to all the pollution and climate change occurring the adaptation of the genome to the environment has been researched. This becomes a serious issue when analyzing how the adaptation will occur within the host. In light of this, the antibody resistance of Pseudomonas has been looked into, in hopes of understanding what adaptation can be done within the host. The highest antibiotic resistance against eight anti- Pseudomonas agents was found amount IN isolates (35-60%), followed by DRC (18-50%) and CH (12-54%) (2). It was also found that the different temperatures and climate conditions affected the antibiotic resistance. Under tropical conditions, between 30 to 37 degrees Celsius, conjugative transfers of antibiotic resistance plasmids occurred more frequently (2). From this research, it can be seen that there is a massive amount of antibiotic resistance (for Pseudomonas) in aquatic ecosystems sediments that have been subjected to contamination (either through waste water or through climate change).
